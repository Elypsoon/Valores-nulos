```python
import pandas as pd
```

```python
df = pd.read_csv('listings - New York.csv')
df.head(10)
```

```python
# pd.set_option('display.max_rows', None)
df.isnull().sum().T
```

```python
# Empezaremos por sustituir los valores nulos de la columna name por 'No name'
# Esto porque es un string único para cada registro y no hay valor por defecto que pueda sustituirlo
df['name'] = df['name'].fillna('No name')
```

```python
df.isnull().sum()
```

```python
# Haremos algo parecido con la columna de description
df['description'] = df['description'].fillna('No description')
df.isnull().sum()
```

```python
# En la columna neighborhood_overview haremos lo mismo
df['neighborhood_overview'] = df['neighborhood_overview'].fillna('No neighborhood_overview')
df.isnull().sum()
```

```python
# Lo mismo con la columna host_location
df['host_location'] = df['host_location'].fillna('No host_location')
df.isnull().sum()
```

```python
# En la columna picture_url, pondremod 'No picture' en lugar de los valores nulos
df['picture_url'] = df['picture_url'].fillna('No picture')
df.isnull().sum()
```

```python
# Con host_name, hacemos lo mismo
df['host_name'] = df['host_name'].fillna('No host_name')
df.isnull().sum()
```

```python
# En la columna host_since, sustituimos los nulos por No date
df['host_since'] = df['host_since'].fillna('No date')
df.isnull().sum()
```

```python
# Con host_about, hacemos lo mismo
df['host_about'] = df['host_about'].fillna('No description')
df.isnull().sum()
```

```python
# Con host_response_time, sustituimos los nulos por 'No data'
df['host_response_time'] = df['host_response_time'].fillna('No data')
df.isnull().sum()
```

```python
# Lo mismo con host_response_rate
df['host_response_rate'] = df['host_response_rate'].fillna('No data')
df.isnull().sum()
```

```python
# Al iguall con host_acceptance_rate
df['host_acceptance_rate'] = df['host_acceptance_rate'].fillna('No data')
df.isnull().sum()
```

```python
# En host_is_superhost, como es un booleano, lo sustituimos por False por defecto
df['host_is_superhost'] = df['host_is_superhost'].fillna("f")
df.isnull().sum()
```

```python
# En host_neighbourhood, sustituimos los nulos por 'No data'
df['host_neighbourhood'] = df['host_neighbourhood'].fillna('No data')
df.isnull().sum()
```

```python
# Con host_thumbnail_url, sustituimos los nulos por 'No picture'
df['host_thumbnail_url'] = df['host_thumbnail_url'].fillna('No picture')
df.isnull().sum()
```

```python
# Lo mismo con host_picture_url
df['host_picture_url'] = df['host_picture_url'].fillna('No picture')
df.isnull().sum()
```

```python
# En host_listings_count, sustituimos los nulos por 1 por defecto, asumiendo que si no hay datos, es porque solo tiene un alojamiento
df['host_listings_count'] = df['host_listings_count'].fillna(1)
df.isnull().sum()
```

```python
# En host_total_listings_count, hacemos lo mismo
df['host_total_listings_count'] = df['host_total_listings_count'].fillna(1)
df.isnull().sum()
```

```python
# En host_verifications, como es un arreglo, sustituimos los nulos por [] para que no pierda la estructura
df['host_verifications'] = df['host_verifications'].fillna('[]')
df.isnull().sum()
```

```python
# En host_has_profile_pic, sustituimos los nulos por 'f' por defecto
df['host_has_profile_pic'] = df['host_has_profile_pic'].fillna('f')
df.isnull().sum()
```

```python
# En host_identity_verified, hacemos lo mismo
df['host_identity_verified'] = df['host_identity_verified'].fillna('f')
df.isnull().sum()
```

```python
# En neighbourhood, sustituimos los nulos por 'No data'
df['neighbourhood'] = df['neighbourhood'].fillna('No data')
df.isnull().sum()
```

```python
# En bathrooms, sustituimos los nulos por 1 por defecto
df['bathrooms'] = df['bathrooms'].fillna(1)
df.isnull().sum()
```

```python
# Con bathrooms_text, sustituimos los nulos por '1 bath'
df['bathrooms_text'] = df['bathrooms_text'].fillna('1 bath')
df.isnull().sum()
```

```python
# Con bedrooms, sustituimos los nulos por 1 por defecto
df['bedrooms'] = df['bedrooms'].fillna(1)
df.isnull().sum()
```

```python
# Con beds, sustituimos los nulos por 1 por defecto
df['beds'] = df['beds'].fillna(1)
df.isnull().sum()
```

```python
# Con price, primero convertimos los valores a numéricos y después sustituimos los nulos con interpolación
df['price'] = df['price'].str.replace('$', '')
df['price'] = df['price'].str.replace(',', '')
df['price'] = pd.to_numeric(df['price'])
df['price'] = df['price'].interpolate()
df.isnull().sum()
```

```python
# Con minimum_minimum_nights, ponemos 0 por defecto
df['minimum_minimum_nights'] = df['minimum_minimum_nights'].fillna(0)
df.isnull().sum()
```

```python
# Con maximum_minimum_nights, minimum_maximum_nights, maximum_maximum_nights, minimum_nights_avg_ntm, maximum_nights_avg_ntm hacemos lo mismo
df['maximum_minimum_nights'] = df['maximum_minimum_nights'].fillna(0)
df['minimum_maximum_nights'] = df['minimum_maximum_nights'].fillna(0)
df['maximum_maximum_nights'] = df['maximum_maximum_nights'].fillna(0)
df['minimum_nights_avg_ntm'] = df['minimum_nights_avg_ntm'].fillna(0)
df['maximum_nights_avg_ntm'] = df['maximum_nights_avg_ntm'].fillna(0)
df.isnull().sum()
```

```python
# Dado que calendar_updated, no contiene ningún valor (toda la columna es nula), la eliminamos
df = df.drop(columns=['calendar_updated'])
df.isnull().sum()
```

```python
# Con has_availability, sustituimos los nulos por 'f' por defecto
df['has_availability'] = df['has_availability'].fillna('f')
df.isnull().sum()
```

```python
# Con first_review y last_review, sustituimos los nulos por 'No date'
df['first_review'] = df['first_review'].fillna('No date')
df['last_review'] = df['last_review'].fillna('No date')
df.isnull().sum()
```

```python
# En review_scores_rating, review_scores_accuracy, review_scores_cleanliness, review_scores_checkin, review_scores_communication, 
# review_scores_location, review_scores_value, sustituimos los nulos por 0
df['review_scores_rating'] = df['review_scores_rating'].fillna(0)
df['review_scores_accuracy'] = df['review_scores_accuracy'].fillna(0)
df['review_scores_cleanliness'] = df['review_scores_cleanliness'].fillna(0)
df['review_scores_checkin'] = df['review_scores_checkin'].fillna(0)
df['review_scores_communication'] = df['review_scores_communication'].fillna(0)
df['review_scores_location'] = df['review_scores_location'].fillna(0)
df['review_scores_value'] = df['review_scores_value'].fillna(0)
df.isnull().sum()
```

```python
# En license, como la gran mayoría de los registros son nulos, eliminamos la columna
df = df.drop(columns=['license'])
df.isnull().sum()
```

```python
# En reviews_per_month, sustituimos los nulos con la interpolación
df['reviews_per_month'] = df['reviews_per_month'].interpolate()
df.isnull().sum()
```

```python
# Verificamos que no haya más valores nulos
df.isnull().sum().sum()
```

```python
# Exportamos el archivo a un nuevo csv con los datos sin nulos
df.to_csv('listings - New York_noNulls.csv')
```
